```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("./feature_labeled.csv", index_col=0)
```

```python
data.describe()
```

```python
label = data.apply(lambda row: 0 if row["vis"] > 0.107638 else 1, axis = 1)
```

```python
data["label"] = label
```

```python
data
```

```python
X = data.iloc[:,[1,2,3]]
Y = data.iloc[:,[4]]
```

```python
data[data["label"] == 0]
```

```python
data[data["label"] == 1]
```

```python
X = X.to_numpy()
Y = Y.to_numpy().T[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, 
                                    test_size=0.3, random_state=42)
```

```python
Y
```

```python
svm_clf = svm.SVC(decision_function_shape='ovo')
svm_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
prediction1 = svm_clf.predict(X_test)
print(accuracy_score(y_test,prediction1),'\n',classification_report(y_test,prediction1))
```
